```python
import plotly.express as px
import pandas as pd
from IPython.display import display

# Load data
perf = pd.read_csv("../data/portfolio_results.csv", parse_dates=["date"])
stats = pd.read_csv("../data/portfolio_stats.csv")

# Show performance by asset type
fig = px.line(
    perf,
    x="date",
    y="cumulative",
    color="asset_type",
    title=" Cumulative Portfolio Performance  Equities vs Crypto",
)
fig.update_layout(template="plotly_dark")
fig.show()

# Optional: Compare daily returns
fig2 = px.box(
    perf,
    x="asset_type",
    y="portfolio_return",
    title=" Distribution of Daily Returns (Equity vs Crypto)",
)
fig2.update_layout(template="plotly_dark")
fig2.show()
```
